```python
df3.info()
```

```python
df3.describe()
```

```python
#Dataframe 행 개수 세기
print(\
    len(df3.loc[(df3['수량'] < 0)])
    )

print(\
        df3.loc[(df3['수량'] < 0)].shape[0]
     )
```

```python
df3.loc[(df3['수량'] < 0)].reset_index()
```

```python
#수량이 -1인 데이터 제거
df3.loc[(df3['수량'] >= 0)].describe()
```

```python
# 수량이 -1인 데이터를 1로 변환
# Step1
condition4 = (df3['수량'] < 0)
df3.loc[condition4, '수량(clean)'] = 1 #condition4를 충족시키는 행에 '수량(Clean)'이라는 필드에 1 할당
df3
```

```python
# 수량이 -1인 데이터를 1로 변환
# Step2
condition4 = (df3['수량'] < 0)

df3['수량(clean)'] = df3['수량'] #'수량(Clean)' 컬럼에 '수량'값 복제
df3.loc[condition4, '수량(clean)'] = 1 #condition4를 충족시키는 행에 '수량(Clean)'이라는 필드에 1 할당
df3
```

```python
df3.describe()
```

```python
condition5 = (df3['상품번호'] == 999999)
df3.loc[condition5]
```

```python
df3['상품번호(clean)'] = df3['상품번호'] #'상품번호(Clean)' 컬럼에 '상품번호'값 복제
condition5 = (df3['상품번호(clean)'] == 999999)
df3.loc[condition5, '상품번호(clean)'] = np.nan #상품번호가 999999인 데이터를 NaN으로 변경
condition6 = (df3['상품번호(clean)'].isnull()) #조건 할당
df3.loc[condition6] #NaN으로 처리한 데이터 확인
```

### JOIN

```python
df7 = pd.read_csv('04_store_member.csv')
print(df7.shape)
df7.head()
```

```python
df8 = pd.read_csv('04_store_product_1.csv')
print(df8.shape)
df8.head()
```

```python
df9 = df8.pivot_table(index = '회원번호', values = ['구매금액', '구매수량'], aggfunc = 'sum').reset_index()
```

```python
pd.merge(df9, df7, on = '회원번호', how = 'inner')
```

```python
pd.merge(df9, df7, on = '회원번호', how = 'right')
```

```python
pd.merge(df8, df7, on = '회원번호', how = 'outer') #pivot table 하지 않은 경우는 데이터 뼝튀기
```

```python
df10 = pd.merge(df9, df7, on = '회원번호', how = 'inner')
```

```python
df10.pivot_table(index = '성별', values = ['구매금액','구매수량'], aggfunc = 'sum')
```
