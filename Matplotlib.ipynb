### 여러 그래프 그리기

```python
x = np.arange(-4.5,5,0.5)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 4*x**2 + 10
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.show()
```

```python
plt.plot(x,y1,x,y2,x,y3)
plt.show()
```

```python
plt.plot(x,y1)# 처음 그리기 함수를 실행하면 그래프 창이 자동으로 생성

plt.figure() #새로운 그래프 창 생성

plt.plot(x,y2)

plt.figure() #새로운 그래프 창 생성

plt.plot(x,y3)

plt.show()
```

```python
#데이터 생성
x = np.arange(0,10,0.1)
y1 = 0.3*(x-5)**2+1
y2 = -1.5*x +3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x)+1

# 2x2 행렬로 이뤄진 하위 그래프에서 P에 따라 위치를 조정
plt.subplot(2,2,1) #p=1
plt.plot(x,y1)

plt.subplot(2,2,2) #p=2
plt.plot(x,y2)

plt.subplot(2,2,3) #p=3
plt.plot(x,y3)

plt.subplot(2,2,4) #p=4
plt.plot(x,y4)

plt.show()
```

```python
x = np.linspace(-4,4,100) #[-4,4] 범위에서 100개의 값 생성
y1 = x**3
y2 = 10*x**2-2

plt.subplot(1,2,1)
plt.plot(x,y1,x,y2) #자동 축

plt.subplot(1,2,2)
plt.plot(x,y1,x,y2)
plt.xlim(-1,1) #x축 확대
plt.ylim(-3,3) #y축 확대
plt.show()
```

### 그래프 꾸미기 

```python
#fmt = ‘[color][line_style][marker]’

x = np.arange(0,5,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3

plt.subplot(2,3,1)
plt.plot(x,y1,x,y2,x,y3,x,y4) #defualt

plt.subplot(2,3,2)
plt.plot(x,y1,'m',x,y2,'y',x,y3,'k',x,y4,'c') #color

plt.subplot(2,3,3)
plt.plot(x,y1,'-',x,y2,'--',x,y3,':',x,y4,'-.') #line_style

plt.subplot(2,3,4)
plt.plot(x,y1,'o',x,y2,'^',x,y3,'s',x,y4,'d') #marker

plt.subplot(2,3,5)
plt.plot(x,y1,'>--r',x,y2,'s-g',x,y3,'d:b',x,y4,'-.Xc') #혼합 사용 

plt.show()
```

### 라벨, 제목, 격자, 범례, 문자열 표시 

```python
x = np.arange(-4.5,5,0.5)
y = 2*x**3

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Graph Title')
plt.grid()

plt.show()
```
