#### Scatter plot
- [연속형] 컬럼 + [연속형] 컬럼 함께 살펴볼 때

디자인 옵션 : 색상, 모양, 크기, 테두리 (color, marker, s, edgecolor)

옵션 : alpha + [hue]

```python
sns.scatterplot(data = df1, x = '나이', y = '마일리지', color = 'skyblue',marker = 'o', s = 150, edgecolor = 'red', alpha = 0.4)
```

```python
df2 = sns.load_dataset('penguins').dropna()
df2.head()
```

```python
sns.scatterplot(data = df2, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species', marker = 'o', s = 150, alpha = 0.4)
```
