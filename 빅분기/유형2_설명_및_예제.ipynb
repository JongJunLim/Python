### Process
1. 원데이터<br>
 1-1. 탐색적 자료 분석(EDA)
    - df.shape : 행/열 확인
    - df.head() : 초기 데이터 확인
    - df.info() : 데이터 정보 확인 (변수명, 데이터 수, 결측치 수, 데이터 타입 등)
    - df.describe() : 기초 통계량 확인 및 이상값 확인
        - train, test 데이터의 데이터 분포 형태를 확인해서 이상치 여부 확인
        - 논리적으로 설명할 수 없는 데이터 (보통 min, max 값에서 확인) (ex : 중고차 생산연도 2060년, 나이 -20, 성별 T 등))
    - df.value_counts() : 범주형 변수 빈도
    - df.isnull().sum() : 결측치 확인
    - df.dropna : 결측치 삭제(행 기준)
        - 전체 데이터 수에 비해 결측값이나 이상값 비율지 작다면 삭제
        - 정보손실이 발생하기 때문에 감안해야 함
    - df['A'].median() / df['A'].mean() / df['A'].mode() : train / test data 둘다 살펴 보기
        - 연속형 : 중앙값 이나 평균값으로 대체
        - 범주형 : 최빈값으로 대체
2. 전처리<br>
3. 전처리 완료<br>
4. Train(80%) / Test(20%) 데이터 분리<br>
 4-1. Train(80%) - Train(60%) : 모델학습 / Validation(20%) : 하이퍼 파라미터 최적화 및 성능평가<br>
 4-2. Test(20%) - 성능 평가의 용도로만 사용됨(Unseen Data)<br>
6. Trained Model <br>
 - Test Data의 X를 넣어 예측된 Y값이 나올 것이고, 이를 실제 Y값(Test Data의 Y값)과 비교
7. Y값 예측<br>
8. 성능 평가<br>
 7-1. 회귀(연속형) : R-Squared, MSE, ...<br>
 7-2. 분류(이산형) : 정확도, 민감도(재현율), 특이도, F1 Score

1. 와인 종류 분류
- 데이터의 결측치, 이상치에 대해 처리하고
- 분류모델을 사용하여 정확도, F1 Score, AUC 값을 산출하시오.

## 4. 모델링 및 성능 평가

```python
# RandomForest (분류 : RandomForestClassifier / 회기 : RandomForestRegressor)
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
# 모델을 사용하여 테스트 데이터 예측
y_pred = model.predict(x_val)
```

```python
# 모델 성능 평가 (정확도, F1 score, 민감도, 특이도 등)
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, recall_score, precision_score
acc = accuracy_score(y_val, y_pred)             #(실제값, 예측값)
f1 = f1_score(y_val, y_pred, average = 'macro') #(실제값, 예측값)
# auc = roc_auc_score(y_val, y_pred)            #(실제값, 예측값) * AUC는 이진 분류(해당 예제는 3개임)
```

```python
# 정확도 (Accuracy)
print("정확도 (Accuracy)")
print(acc)
print("\n")
print("---")
print("\n")
print("macro f1 score")
print(f1)
```

## (주의) test 셋을 모델에 넣어 나온 예측값을 제출해야함

```python
# (실기시험 안내사항)
# 아래 코드 예측변수와 수험 번호를 개인별로 변경하여 활용
# pd.DataFrame({ 'result' : y_result }).to_csv('수험번호.csv', index = False)

# 모델을 사용하여 테스트 데이터 예측

# 1. 특정 클래스로 분류할 경우 (predict)
y_result = model.predict(x_test)
print(y_result[ : 5])
print("\n")
print("---")
print("\n")

# 2. 특정 클래스로 분류될 확률을 구할 경우 (predict_proba)
y_result_prob = model.predict_proba(x_test)
print(y_result_prob[ : 5])
print("\n")
print("---")
print("\n")

# 이해해보기
result_prob = pd.DataFrame({
    'result' : y_result,
    'prob_0' : y_result_prob[:,0],
    'prob_1' : y_result_prob[:,1],
    'prob_2' : y_result_prob[:,2]
})

# Class 0 일 확률 : y_result_prob[:,0]
# Class 1 일 확률 : y_result_prob[:,1]
# Class 2 일 확률 : y_result_prob[:,2]

print(result_prob[:5])
```
