Q57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 <br>
Unknown : N<br>
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e

```python
df6['Income_Category'].unique()
```

```python
dic = {
        '$60K - $80K' : 'c', 
        'Less than $40K' : 'a',
        '$80K - $120K' :'d', 
        '$40K - $60K' : 'b',
        '$120K +' : 'e',
        'Unknown' : 'N'
      }
dic
```

```python
#map 함수 : mapping 한다고 생각하면 됨
df6.loc[ : , 'new_income'] = df6['Income_Category'].map(dic)
df6.head()
```

$40K - $60K : b
$60K - $80K : c

```python
def date_ft(x) :
    if x[-2 : ] != '24' :
        x = x.replace( ':' , ' ')
        return pd.to_datetime(x)
    else:
        date = pd.to_datetime(x.split(':')[0]) + datetime.timedelta( days= 1)
        return date
```
