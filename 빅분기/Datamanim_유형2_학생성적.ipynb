```python
import pandas as pd
import numpy as np
```

```python
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_train.csv")
x_test = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/X_test.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/studentscore/y_train.csv")

# 데이터 확인
print(x_train.head())
print("---")
print("\n")
print(x_test.head())
print("---")
print("\n")
print(y_train.head())
```

```python
# 데이터 프레임 형태 확인 
print(x_train.info()) # object = ['school', 'sex','address','famsize','Pstatus','Mjob', 'Fjob','reason','guardian','schoolsup','famsup','paid','activities']
print("---")
print("\n")
print(x_test.info())
print("---")
print("\n")
print(y_train.info())
```

```python
# 기술통계량 확인
print(x_train.describe(include=['int','float','object']))
print("---")
print("\n")
print(x_test.describe(include=['int','float','object']))
print("---")
print("\n")
print(y_train.describe(include=['int','float','object']))
```

```python
# 결측치 확인
print(x_train.isnull().sum())
print("---")
print("\n")
print(x_test.isnull().sum())
print("---")
print("\n")
print(y_train.isnull().sum())
```

```python
# 변수처리
StudentID = x_test['StudentID'].copy()

x_train = x_train.drop(columns=['StudentID'])
x_test = x_test.drop(columns=['StudentID'])

# 원핫인코딩
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)

print(x_train.info())
print(x_test.info())
```

```python
# 검증
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train['G3'],test_size = 0.2, random_state = 2023)

print(x_train.shape)
print("---")
print(x_val.shape)
print("---")
print(y_train.shape)
print("---")
print(y_val.shape)
```

```python
# 모델 적용 및 평가
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state = 2023)

print(x_train.info())
print(x_test.info())
print(y_train.info())
```

```python
# 랜덤 포레스트 모델 사용 (참고 : 분류모델은 RandomForestClassifier)
model.fit(x_train, y_train)
```

```python
# 모델을 사용하여 테스트 데이터 예측
y_pred = model.predict(x_val)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_val, y_pred) #실제값, 예측값
r2 = r2_score(y_val, y_pred) #실제값, 예측값

# MSE
print("MSE")
print(mse)
print("---")
# R2 Score
print("R-Squared")
print(r2)
# RMSE
rmse = mse**0.5
print("---")
print("RMSE")
print(rmse)
```

```python
# 실기시험 안내사항
# 아래 코드 예측변수와 수험번호를 개인별로 변경하여 활용
# pd.DataFrame({'StudentID' : StudentID, 'G3' : y_result}).to_csv('수험번호.csv', index = False)

# 모델을 사용하여 테스트 예측
y_result = model.predict(x_test)
result = pd.DataFrame({'StudentID' : StudentID, 'G3' : y_result})
print(result.head(10))
```
