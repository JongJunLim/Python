Q12. hp 변수에 대해 데이터 표준화(Z-score)를 진행하고, 이상치의 수를 구하시오<br>
(단, 이상치는 Z 값이 1.5를 초과하거나 -1.5 미만인 값이다) <br>
<br>
Z = (X - 평균)/ 표준편차

```python
mean = df['hp'].mean()
std = df['hp'].std()
```

```python
df['zcore'] = (df['hp'] - mean) / std
df[['hp', 'zcore']].head()
```

```python
len(df.loc[(df['zcore'] < -1.5) | (df['zcore'] > 1.5)])
```

Q13. mpg 컬럼을 최소최대 Scaling을 진행한 수 0.7보다 큰 값을 가지는 레코드 수를 구하라<br>
<br>
(x - min) / (max - min)

```python
#방식1
df['mpg_Scaling'] = MinMaxScaler().fit_transform(df[['mpg']])
```

```python
df.head()
```

```python
print(len(df.loc[ df['mpg_Scaling'] > 0.7]))
```

```python
#방식2
min_mpg = df['mpg'].min()
max_mpg = df['mpg'].max()
df['mpg_scaling2'] = ( df['mpg'] - min_mpg) / (max_mpg - min_mpg)
```

```python
df.head()
```

```python
print(len(df.loc[ df['mpg_scaling2'] > 0.7]))
```

Q14. wt 컬럼에 대해 상자그림 기준으로 이상치의 개수를 구하시오 <br>

outlier : IQR * 1.5 이상

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
Q1 = df['wt'].quantile(0.25)
Q3 = df['wt'].quantile(0.75)
IRQ = Q3 -Q1 

upper = Q3 + 1.5 * IRQ
lower = Q1 - 1.5 * IRQ

print(len(df.loc[(df['wt'] < lower) | (df['wt'] > upper)]))
```
